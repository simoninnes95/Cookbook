```python
# Contains Duplicate
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:

        count = {}

        for i in range(0, len(nums)):
            if nums[i] in count:
                count[nums[i]] += 1
            else: 
                count[nums[i]] = 1

        for item in list(count.values()):
            if item > 1:
                return True
        
        return False
        
```

```python
nums = [1,2,2,3]

s = Solution()
s.hasDuplicate(nums)
```
